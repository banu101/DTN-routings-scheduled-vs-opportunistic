# Way To Graduation

In the DTN cimulator, if we set N number of nodes in the network, it automatically adds source and destination nodes. Thus we have N+2 nodes in the network. 
1. N = 0 node is the source node.
2. N = 1 is the destination node.
3. Other nodes are moving nodes in the network.
4. The numbber of a node coressponds to the same raw and coloumn number in the Excel sheet.

We convert the data from dataframe to list of lists, so it's easier to use. 

```python
m = 0;

dataNew = [];


for i in range(N):
    for j in range(N):
        if(type(data[j][i]) == str):
                string = data[j][i].split(",")
                integerList =[eval(m) for m in string]
                dataNew.append(integerList)
        else:
            dataNew.append([])

dataNew = np.reshape(dataNew, (N, N)).T

print(dataNew)
```

### Brute Force Algorithm

```python
class Node:
 
    def __init__(self, time, transm):
        self.transm = transm
        self.time = time
        

#We make N nodes with time when transmission happened and the node that transmitted

node = [0]*N
node[0] = Node(0, 0)  #Node 0 is the source node

for i in range(1, N):
    node[i] = Node(100000, i)
```

```python
#We find all the time interval start values, and place them in the increasing order, and we construct dictoionary
#where keys are time values
dataDict = {}
for i in range(N):
    for j in range(i+1, N):
        for k in dataNew[j][i]:
            if(k in dataDict):
                dataDict[k].append([i, j])
            else:
                dataDict.update({k: [[i, j]]})
           
keyList = list(dataDict.keys())
keyList.sort()
print(keyList)
dataDict
```

```python
#We update the nodes
tr = 0

for i in keyList:
    for j in range(len(dataDict[i])):
            if((node[dataDict[i][j][0]].time != 100000) and (node[dataDict[i][j][1]].time == 100000)):
                node[dataDict[i][j][1]] = Node(i, dataDict[i][j][0])
                tr = tr + 1
            elif((node[dataDict[i][j][1]].time != 100000) and (node[dataDict[i][j][0]].time == 100000)):
                node[dataDict[i][j][0]] = Node(i, dataDict[i][j][1])
                tr = tr + 1

for i in range(len(node)):
    print(i, end="")
    print(': ', end="")
    print(node[i].time, end="")
    print(' (', end="")
    print(node[i].transm, end="")
    print(')\n')
```

The results above show the node numbers on the left, time of transmission in the middle, and the node that has transmitted on the right. 

```python
#Print the shortest path and time at which transmissions happened

sPathList = []
i = 1;

while (i != 0):
    sPathList.append([i, node[i].time, node[i].transm])
    i = node[i].transm
    

sPath = pd.DataFrame(sPathList)
sPath.columns = ["Node", "Time", "Transmitter"]
print(sPath, "\n")
print("The number of shortest path transmissions: ", len(sPathList))
print("The number of overall transmissions: ", tr)
```

In the table above, the first row is the destination source, and the second row contains the node that transmitted the package to the destination source. In the same manner, we have all the nodes which are included in teh shortest path.
